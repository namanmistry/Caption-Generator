```python
import numpy as np
import pandas as pd
import cv2
import os
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from keras.utils.np_utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
incept_model = ResNet50(include_top=True)
last = incept_model.layers[-2].output
modele = Model(inputs=incept_model.input, outputs=last)
```

```python
image_features = {}
counter = 0
for image in os.listdir("./data/Images"):
    img = cv2.imread(f"./data/Images/{image}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224,224))

    img = img.reshape(1,224,224,3)
    pred = modele.predict(img).reshape(2048,)

    image_features[image] = pred
    counter +=1

    if counter>2000:
        break
    elif counter % 50 ==0:
        print(counter)
```

```python
caption_path = "./data/captions.txt"
captions = open(caption_path, 'rb').read().decode('utf-8').split('\n')

caption_dic = {}
for i in captions:
    try:
        img_name = i.split(',')[0]
        caption = i.split(',')[1]
        if img_name in image_features:
            if img_name not in caption_dic:
                caption_dic[img_name] = [caption]
            else:
                caption_dic[img_name].append(caption)
    except:
        pass
```

```python
def preprocessed(txt):
    modified = txt.lower()
    modified = "startofseq " + modified + " endofseq"
    return modified

for k,v in caption_dic.items():
    for vv in v:
        caption_dic[k][v.index(vv)] = preprocessed(vv)
```

```python
count_words = {}
count = 1
for k,vv in caption_dic.items():
    for v in vv:
        for word in v.split():
            if word not in count_words:
                count_words[word] = count
                count += 1
count_words
```

```python
for k,vv in caption_dic.items():
    for v in vv:
        encoded = []
        for word in v.split():
            encoded.append(count_words[word])

        caption_dic[k][vv.index(v)] = encoded
```

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
MAX_LEN = 0
for k, vv in caption_dic.items():
    for v in vv:
        if len(v)> MAX_LEN:
            MAX_LEN = len(v)
            print(v)
MAX_LEN
```

```python
VOCAB_SIZE = len(count_words)

def generator(photo, caption):
    n_samples = 0
    
    X = []
    y_in = []
    y_out = []
    
    for k, vv in caption_dic.items():
        for v in vv:
            for i in range(1, len(v)):
                X.append(photo[k])
                in_seq = [v[:i]]
                out_seq = v[i]
                
                in_seq = pad_sequences(in_seq, maxlen=MAX_LEN, padding='post', truncating='post')[0]
                out_seq = to_categorical([out_seq], num_classes=VOCAB_SIZE+1)[0]
                
                y_in.append(in_seq)
                y_out.append(out_seq)
                
    return X, y_in, y_out
```

```python
X, y_in, y_out = generator(image_features, caption_dic)
```

```python
X = np.array(X)
y_in = np.array(y_in, dtype='float64')
y_out = np.array(y_out, dtype='float64')
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.layers import Dense, Flatten,Input, Convolution2D, Dropout, LSTM, TimeDistributed, Embedding, Bidirectional, Activation, RepeatVector,Concatenate
from tensorflow.keras.models import Sequential, Model
```

```python
embedding_size = 128
max_len = MAX_LEN
vocab_size = len(count_words)+1

image_model = Sequential()

image_model.add(Dense(embedding_size, input_shape=(2048,), activation='relu'))
image_model.add(RepeatVector(max_len))

image_model.summary()

language_model = Sequential()

language_model.add(Embedding(input_dim=vocab_size, output_dim=embedding_size, input_length=max_len))
language_model.add(LSTM(256, return_sequences=True))
language_model.add(TimeDistributed(Dense(embedding_size)))

language_model.summary()

conca = Concatenate()([image_model.output, language_model.output])
x = LSTM(128, return_sequences=True)(conca)
x = LSTM(512, return_sequences=False)(x)
x = Dense(vocab_size)(x)
out = Activation('softmax')(x)
model = Model(inputs=[image_model.input, language_model.input], outputs = out)

# model.load_weights("../input/model_weights.h5")
model.compile(loss='categorical_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
model.summary()
```

```python
model.fit([X, y_in], y_out, batch_size=512, epochs=100)
```

```python
inv_dict = {v:k for k, v in count_words.items()}
```
